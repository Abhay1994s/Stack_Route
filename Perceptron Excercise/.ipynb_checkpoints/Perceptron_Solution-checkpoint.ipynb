# PERCEPTRON BASIC EXERCISES

## Importing necessary modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib import animation
%matplotlib nbagg
#%matplotlib inline
import math
```

## Importing files

```python
data_set_1 = pd.read_csv("datafile1.csv", names = ["X1","X2","Y"])
data_set_2 = pd.read_csv("datafile2.csv", names = ["X1","X2","Y"])
data_set_3 = pd.read_csv("datafile3.csv", names = ["X1","X2","Y"])
data_set_4 = pd.read_csv("datafile4.csv", names = ["X1","X2","Y"])
```

# First DataSet

```python
plt.scatter(data_set_1['X1'], data_set_1['X2'], c = data_set_1['Y'])
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Feature Plot X1 vs X2')
plt.show()
```

```python
## This Plot can be linearly separable
```

```python
x_values = []
y_actual = []
for i in data_set_1.iterrows():
    x1 = i[1][0]
    x2 = i[1][1]
    y = i[1][2]
    x_values.append(np.array([x1,x2,1]))
    y_actual.append(np.array(y).reshape(1))
x_values[0]
```

```python
# Estimate Perceptron weights using stochastic gradient descent
def train_weights(train, l_rate, n_epoch):
    weights_plot = []
    w1 = np.random.random_sample() # weight 1 coresponding to X1
    w2 = np.random.random_sample() # weight 2 coresponding to X2
    bias = np.random.random_sample() # bias
    for epoch in range(n_epoch):
        sum_error = 0.0
        weights_plot.append(np.array([bias,w1,w2]).reshape(3,1))
        for i in data_set_1.iterrows():
            X_1 = i[1][0]
            X_2 = i[1][1]
            Y_actual = i[1][2]
            #print("Actual X1 is: ", X_1 ,", Actual X2 is: " ,X_2, ",Actual Y is: ", Y_actual)
            activation_value = (w1*X_1)+(w2*X_2)+bias
            if activation_value >=0:
                Y_predicted = 1
            else:
                Y_predicted = 0
            error = Y_actual - Y_predicted
            sum_error += error**2
            w1 = w1 + (error*X_1*l_rate)
            w2 = w2 + (error*X_2*l_rate)
            bias = bias + error
        print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error))
    return bias,w1,w2,weights_plot
```

```python
# Calculate Weights
l_rate = 0.1
n_epoch = 20
bias_1, w1_1, w2_1, weights_plot_1 = train_weights(data_set_1, l_rate, n_epoch)
print(bias_1,w1_1,w2_1)
```

```python
# Equation of a line
#-(bias1_1/w2_1) + -((w1_1*x1)/w2_1) = x2
```

```python
plt.plot(data_set_1['X1'], -(bias_1/w2_1) -((w1_1*data_set_1['X1'])/w2_1), color='red')
plt.scatter(data_set_1['X1'], data_set_1['X2'], c = data_set_1['Y'])
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Feature Plot X1 vs X2')
plt.show()
```

### Animated Plot

```python
#First set up the figure, the axis, and the plot element we want to animate

fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

ax.scatter(data_set_1['X1'], data_set_1['X2'], c = data_set_1['Y'])
# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def animate(i):
    x = data_set_1['X1']
    y = -(weights_plot_1[i][0]/weights_plot_1[i][2]) -((weights_plot_1[i][1]*x)/weights_plot_1[i][2])
    line.set_data(x, y)
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=len(weights_plot_1), interval=20, blit=True,repeat = False)
plt.show()
```
